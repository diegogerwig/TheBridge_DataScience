## Ejercicio 2
Escribe un programa para pasar de grados a radianes. Hay que usar `input`. Recuerda que la conversión se realiza mediante

$$radianes = grados*(pi/180)$$

````python
import math
grados = input('Cuantos grados quieres convertir a radianes? :')
radianes = float(grados) * (math.pi / 180)
print(radianes)
````

## Ejercicio 6
Mismo programa que en 5, pero en esta ocasión tienen que ser tres inputs y dos salidas. Una de las salidas que nos indique si todos son iguales, y la otra si al menos dos inputs sí que lo son

````python
input1 = input('Escribe algo para el input 1: ')
input2 = input('Escribe algo para el input 2: ')
input3 = input('Escribe algo para el input 3: ')
if (input1 == input2 and input2 == input3):
    print('Los 3 inputs son iguales')
elif(input1 == input2 or input1 == input3 or input2 == input3):
    print('Hay 2 inputs iguales')
else:
    print('Todos los inputs son distintos')
````

## Ejercicio 7
Crea un programa que recoja dos inputs. Tiene que comprobar si su suma es igual, superior o inferior a 10, cada condición por separado

````python
try:
  input1 = float(input('Escribe un numero: '))
  input2 = float(input('Escribe otro número: '))
  resultado = input1 + input2
  if resultado == 10:
      print('La suma es 10')
  elif resultado > 10:
      print('La suma es mayor a 10')
  else:
      print('La suma es menor a 10')

except ValueError:
        print('Por favor, asegúrate de ingresar números válidos.')
````

## Ejercicio 10
Para el siguiente string se pide imprimir por pantalla los siguientes casos

    "Lo primero de todo, cómo están los máquinas"

1. Pasarlo todo a mayúsculas
2. Pasarlo todo a minúsculas
3. Pasar a mayúsculas solo las iniciales de las palabras
4. Crea una lista dividiendolo por sus espacios
5. Sustituye las comas `,` por puntos y comas `;`
6. Elimina las `a` minúsculas

````python
my_str = "Lo primero de todo, cómo están los máquinas"
print(my_str.upper())
print(my_str.lower())
print(my_str.title())
print(my_str.split(' '))
print(my_str.replace(',', ';'))
print(my_str.replace('a', ''))
````
