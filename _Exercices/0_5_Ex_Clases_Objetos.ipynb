## Ejercicio 1
Somos una empresa que tiene varias tiendas de electrodomésticos. Necesitamos un programa para manejar las tiendas, ver las ventas que han tenido, cuántos empleados hay, etc... Para ello vamos a modelizar la tienda en una clase, que tendrá las siguientes características:
* Nombre clase: 'Tienda'
* Atributos comunes:
    * Tipo: 'Electrodomésticos'
    * Abierta: True
* Atributos propios de cada tienda:
    * Nombre - String
    * Dirección - String
    * Número de empleados - int
    * Ventas últimos 3 meses - Lista de 3 numeros
* Método para calcular las ventas de todos los meses, que devuelva un número con todas las ventas.
* Método que calcula la media de ventas de los últimos meses, por empleado
* Método que devuelve en un string el nombre de la tienda, junto con su dirección.
* Método que obtiene las ventas del último mes.
* Método para dar una proyección de las ventas en caso de haber invertido X dinero en marketing en esa tienda. Siendo X el parámetro de entrada. Si X es menor de 1000, las ventas de los ultimos 3 meses hubiesen sido de (1.2 x ventas), si es mayor o igual de 1000, las ventas hubiesen sido de (1.5 x venta). El método reescribe el atributo de ventas en los últimos 3 meses, y además devuelve las ventas con el aumento.

Se pide:
1. Implementar la clase 'Tienda'
2. Crear tres tiendas con datos inventados
3. Comprueba en al menos una de ellas, todo lo implementado en la clase tienda (sus atributos, media de ventas, ventas/empleado...)
3. Calcular las ventas del último mes de todas las tiendas. Para ello usa el bucle `for`
4. Imprime por pantalla los nombres de las tiendas cuya dirección lleve el string 'Avenida'
5. Seguro que ya has hecho este apartado... Documenta la clase :)

```python
class Tienda:
    '''
    Info de una tienda
    '''

    tipo = 'electrodomesticos'
    abierto = True

    def __init__(self, nombre:str, direc:str, empl:int, ventas:list):
        '''
        nombre: nombre de la tienda
        direc: direccion postal
        empl: numero de empleados
        ventas: importe de las ventas de los 3 últimos meses
        '''
        self.nombre = nombre
        self.direc = direc
        self.empl = empl
        self.ventas = ventas

    def ventas_trimestre(self):
        return sum(self.ventas)

    def prom_ventas(self):
        return ((sum(self.ventas)/len(self.ventas)) / self.empl)

    def ubicacion(self):
        return 'La tienda ' + self.nombre + ' se encuentra situada en ' + self.direc

    def ventas_mes(self):
        return sum(self.ventas) / 3
    
    def proyec_ventas(self, inv_mark:int):
        if inv_mark < 1000:
            factor = 1.2
        else:
            factor = 1.5
        ventas_plus = [venta * factor for venta in self.ventas]
        self.ventas = ventas_plus
        return ventas_plus
```

```python
tienda1 = Tienda('ELECT ANA', 'Avenida Libertad', 123, [10, 20, 30])
tienda2 = Tienda('ELECT BEA', 'Calle Tierra', 12, [40, 50, 60])
tienda3 = Tienda('ELECT CARLOS', 'Avenida Mar', 23, [10, 20, 30])
print(f'DOC Tienda 1: {tienda1.__doc__}')
print(f'DOC Tienda 1: {tienda1.__init__.__doc__}')
print(f'DICT Tienda 1: {tienda1.__dict__}')
print(f'DICT Tienda 2: {tienda2.__dict__}')
print(f'DICT Tienda 3: {tienda3.__dict__}')
print(f'Ventas trimestre: {tienda1.ventas_trimestre()}')
print(f'Promedio ventas: {tienda1.prom_ventas()}')
print(f'Ubicación: {tienda1.ubicacion()}')
print(f'Ventas mes: {tienda1.ventas_mes()}')
print(f'Inversión en marketing -> 999 €: {tienda1.proyec_ventas(999)}')
print(f'Inversión en marketing -> 1001 €: {tienda3.proyec_ventas(1001)}')
```

## Ejercicio 2
En este ejercicio vamos a implementar una clase *Perro* en Python. La clase tiene las siguientes características:
* Cosas que sabemos seguro que tiene un perro
    * Tiene 4 patas
    * 2 orejas
    * 2 ojos
    * Una velocidad de 0. Por defecto, el perro está parado
* Cuando se inicialice:
    * El perro será de una determinada raza
    * Por defecto tendrá pelo 'Marrón', a no ser que se diga lo contrario.
    * Por defecto no tendrá dueño, a no ser que se diga lo contrario.
    
* Dispondrá también de un método llamado andar, que tiene un argumento de entrada (aumento_velocidad). Este valor se le sumará a la velocidad que ya llevaba el perro.
* Necesita otro método (parar), donde pondremos la velocidad a 0.
* Otro método llamado 'ladrar', que tendrá un argumento de entrada, y la salida será el siguiente string: 'GUAU!' + el argumento de entrada.


Se pide:
* Implementa la clase *Perro*
* Crea un objeto de tipo *Perro*, sin dueño
* Comprueba que están bien todos sus atributos
* Prueba que ande, y comprueba su velocidad
* Páralo
* Documenta la clase *Perro*

```python
class Perro:
    '''
    Clase que representa a un perro.
    
    Atributos:
        raza: La raza del perro.
        color_pelo: El color del pelo del perro.
        amo: El dueño del perro.
        velocidad: La velocidad actual del perro.
    '''
    def __init__(self, raza:str, color_pelo:str='Marrón', amo:str=None):
        self.raza = raza
        self.color_pelo = color_pelo
        self.amo = amo
        self.velocidad = 0

    def andar(self, aumento_velocidad:int):
        ''' 
        Aumenta la velocidad del perro.
        '''
        self.velocidad += aumento_velocidad

    def parar(self):
        '''
        Detiene al perro, poniendo su velocidad a 0.
        '''
        self.velocidad = 0

    def ladrar(self, mensaje:str):
        '''
        Hace que el perro ladre.
        '''
        return 'GUAU! ' + mensaje
```

```python
perro_1 = Perro('Labrador')
print(f'DOC Perro 1: {perro_1.__doc__}')
print(f'DICT Perro 1: {perro_1.__dict__}')

print('Raza:', perro_1.raza)
print('Color de pelo:', perro_1.color_pelo)
print('Dueño:', perro_1.amo)
print('Velocidad:', perro_1.velocidad)

perro_1.andar(10)
print('Velocidad después de andar:', perro_1.velocidad)

perro_1.parar()
print('Velocidad después de parar:', perro_1.velocidad)

print(perro_1.ladrar('¡Hola!'))

print(30 * '*')

perro_2 = Perro('Braco', 'Black', 'San Roque')
print(f'DOC Perro 2: {perro_2.__doc__}')
print(f'DICT Perro 2: {perro_2.__dict__}')
print('Velocidad:', perro_2.velocidad)
perro_2.andar(22)
print('Velocidad después de andar:', perro_2.velocidad)
print(f'DICT Perro 2: {perro_2.__dict__}')
print('Velocidad:', perro_2.velocidad)
perro_2.andar(11)
print('Velocidad después de andar:', perro_2.velocidad)
print(f'DICT Perro 2: {perro_2.__dict__}')
print('Velocidad:', perro_2.velocidad)
perro_2.parar()
print('Velocidad después de parar:', perro_2.velocidad)
print(f'DICT Perro 2: {perro_2.__dict__}')
print('Velocidad:', perro_2.velocidad)
```
